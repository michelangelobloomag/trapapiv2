### This script downloads the Spanish data for weather stations ###

depends from:

* ClimateNormals_spain.pickle.gz

in order to have same set of points downloaded every time


NOTE. there is another code supposed to be used to initialize the climate data

```python
#!/usr/bin/env python
# coding: utf-8


import pandas as pd
import pickle
import time


# Import Meteostat library and dependencies
from datetime import datetime
from datetime import date
from datetime import timedelta

import matplotlib.pyplot as plt
from meteostat import Point as meteo_Point
from meteostat import Daily as meteo_Daily
```

```python



def getLocationPoints():
    ## get the climateNormals coord points
    ## TODO:
    ## make this function parametric in directory + name of the pest
    
    cmdf = pd.read_pickle("savedModels/spodoptera/ClimateNormals_spain.pickle.gz")
    cmf = cmdf.loc['12-01']
    vector_of_points = zip(cmf.groupby('pnum').lat.apply(list),cmf.groupby('pnum').lon.apply(list),
                        cmf.groupby('pnum').elevation.apply(list), cmf.pnum.unique())
        
    return vector_of_points



def getWeatherActual(vector_of_points,start,stop):
    ## this function downloads and averages the last 15 years of data for historical normals
    allWeatherData = []
    
        
    for i, p in enumerate(vector_of_points): 
    
        lat = p[0][0]
        lon = p[1][0]
        elev = p[2][0]
        pnum = p[3]
        end = stop
        point_meteostat = meteo_Point(lat,lon)
        data = []
        print(lat,lon,elev,pnum)
        try:
            try:
                data = meteo_Daily(point_meteostat, start, end)
                df = data.fetch()
                
                df = df.assign(pnum = pnum)
                df = df.assign(elev = elev)
                df = df.assign(lat = lat)
                df = df.assign(lon = lon)
                allWeatherData.append(df)
                
            except Exception as e:
                print(str(e))
                pass
        except:
            pass
       
        if(i%10 == 0 and pnum >0):
            pickle.dump(allWeatherData, open("allWeather_tmp_spain.pickle","wb")) ## tmp file
            print("Done", i)
        time.sleep(2)  
        
    pickle.dump(allWeatherData, open("allWeather_tmp_spain.pickle","wb"))
        
    return pd.concat(allWeatherData)



```

```python

 
# Get today's date
today = datetime.today()
 
# Yesterday date
yesterday = today - timedelta(days = 1)


oneweekago = today - timedelta(days = 7)
start = oneweekago

###start = datetime(2023,6,1) ### temporary (remove it )
stop = yesterday


vector_of_points = getLocationPoints()
weatherDf = getWeatherActual(vector_of_points,start,stop )


weatherDf.to_pickle("WeatherAndClimateData/WeatherSpain__%s__%s.pickle.gz"%(start,stop))
```
